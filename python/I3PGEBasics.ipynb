```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
I3MuonEffectiveArea = np.load("../muon_effective_area/muon_eff_area.pickle")
```

```python
I3MEA = I3MuonEffectiveArea['eff_muon_area']
energy_bins = I3MuonEffectiveArea['energy_bins']
costh_bins = sorted(I3MEA.keys(),key=lambda cospair: cospair[0])
energy_bins_center = [(energy_bins[i]+energy_bins[i+1])/2. for i in range(len(energy_bins)-1)]
costh_bins_center = [(costh_bin[0]+costh_bin[1])/2. for costh_bin in costh_bins]
```

```python
plt.figure(figsize=(8,6))

for i,costh_bin in enumerate(costh_bins):
    if i%3 != 0:
        continue
    plt.hist(energy_bins_center, 
             weights = I3MEA[costh_bin], bins = energy_bins, histtype = "step",
             lw = 2)
    
plt.semilogx()
```
