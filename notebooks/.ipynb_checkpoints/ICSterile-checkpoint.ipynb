```python
def MakeContourPlot(filepath,flv):
    file = open(filepath,'r')
    data = np.array(gt.quickread(file))
    file.close()
    
    enu = np.log10(data[:,1]/(1.0e9))
    costh = data[:,0]
    nuf = data[:,2+flv]
    
    enu_min = min(enu)
    enu_max = max(enu)
    costh_min = min(costh)
    costh_max = max(costh)
    step_enu = (enu_max - enu_min)/500.0
    step_costh = (costh_max - costh_min)/100.0
    nenu = np.arange(enu_min,enu_max,step_enu)
    ncosth = np.arange(costh_min,costh_max,step_costh)
    
    nnuf = pl.griddata(costh,enu,nuf,ncosth,nenu)#,interp = 'linear')
    
    plt.figure(figsize = (9,7))
    #plt.semilogy()
    plt.ylim(2.0,4.0)
    plt.contourf(ncosth,nenu,nnuf,100)
    plt.colorbar()
```

```python
def MakeContourPlotRatio(filepath_in,filepath_out,flv):
    file = open(filepath_in,'r')
    data = np.array(gt.quickread(file))
    file.close()
    #enu = data[:,1]/(1.0e9)
    enu = np.log10(data[:,1]/(1.0e9))
    costh = data[:,0]
    nuf_in = data[:,2+flv]
    
    file = open(filepath_out,'r')
    data = np.array(gt.quickread(file))
    file.close()
    nuf_out = data[:,2+flv]
    
    ratio = nuf_out/nuf_in
    
    enu_min = min(enu)
    enu_max = max(enu)
    costh_min = min(costh)
    costh_max = max(costh)
    step_enu = (enu_max - enu_min)/500.0
    step_costh = (costh_max - costh_min)/100.0
    
    nenu = np.arange(enu_min,enu_max,step_enu)
    ncosth = np.arange(costh_min,costh_max,step_costh)
    #nratio = pl.griddata(costh,enu,ratio,ncosth,nenu,interp = 'linear')
    nratio = pl.griddata(costh,enu,ratio,ncosth,nenu)
    
    plt.figure(figsize = (10,8))
    plt.ylim(2,6)
    #plt.ylim(1.0e2,1.0e6)
    #plt.semilogy()
    plt.contourf(ncosth,nenu,nratio,100)
    v = np.linspace(0.0, 1.0, 11, endpoint=True)
    plt.colorbar(ticks = v)

    #plt.clim(0,1)
    
    plt.xlabel(r"$\cos \theta_z$")
    plt.ylabel(r"$\mathrm{log}_{10}(E/\mathrm{GeV})$")
```
