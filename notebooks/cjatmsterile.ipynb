```python
def EnuDistribution(i):
    figure(figsize = (8,6))
    loglog()
    xlabel(r'$E_\nu$')
    ylabel("Events")
    plot(enu_ar,split(std,len(costh_ar))[i][0])
    
def EnuDistribution():
    figure(figsize = (12,8))
    loglog()
    xlabel(r'$E_\nu$')
    ylabel("Events")
    
    events = split(std,len(costh_ar))[0][0]
    for j in range(1,len(costh_ar)):
        events += split(std,len(costh_ar))[j][0]
    n, bins, patches = plt.hist(enu_ar,bins = enu_ar,
                                weights = events, histtype = 'step')
    #plt.ylim(1.0e-2,1.0e3)
    
def CosthDistribution():
    figure(figsize = (8,6))
    xlabel(r'$\cos (\theta_{z})$')
    ylabel("Events")
    n, bins, patches = plt.hist(costh_ar,bins = costh_ar,
                                weights = map(sum,std), histtype = 'step')
```
